```python
# Mean of total Sales from each state
column = df1["Total in state"][:-1]
average = np.mean(column)
print(f"Mean sales of all the States combined is : {average}")
```

```python
y = df1["Total in state"][:30]
x = df1["State"][:30]

plt.bar(x, y)
plt.xlabel("State")
plt.ylabel("Total EV's in state")
plt.axhline(y.mean(), color='r', linestyle='--')
#plt.axhline(y.median(), color='g', linestyle='--')
plt.title("State V/S Total EV's in state")
plt.xticks(rotation=90)
plt.show()
```

Horizonal red-dotted line represents the mean....so the state having the sales above mean have higher chance of increased sales in the upcoming year as well


```python
states_with_greater_sales = df1["State"][:30][df1["Total in state"] > average]
states_with_greater_sales
```

We can see Maharashtra, Gujrat and Uttar Pradesh has the highest registered EV sales
and Manipur, Meghalaya and  Himachal Pradesh has lowest EV registered

Some Visualizations regarding the data

```python
sns.countplot(data=df3, x="Profession", palette="flare")
plt.title("Profession data")
plt.xlabel("profession")
plt.ylabel("Count")
plt.show()
```

```python
# Histogram of Ages
sns.histplot(data=df3, x="Age", bins=10, color="yellow", edgecolor="blue")
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Numbers")
plt.show()
```

```python
# Marital Status
sns.countplot(data=df3, x="Marrital Status", palette="dark")
plt.title("Marrital Status Analysis")
plt.xlabel("Marrital Status")
plt.ylabel("Count")
plt.show()
```

```python
# Salary Analysis
sns.kdeplot(df3["Salary"])

plt.title("Salary Distribution")
plt.xlabel("Salary")
plt.ylabel("Density")

plt.show()
```

Car Make Comparison

```python
sns.countplot(data=df3,x="Make")
plt.title("Car Make Comparison")
plt.xlabel("Make")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.show()
```

```python
# Finding Correlation matrix
data_corr = df3.corr()

plt.figure(figsize=(8,6))
plt.title("Correlation Matrix HeatMap")

sns.heatmap(data_corr, annot=True, fmt=".2f", cmap="inferno", linewidth=0.5)

plt.xticks(rotation=45)
plt.yticks(rotation=0)

plt.show()
```
